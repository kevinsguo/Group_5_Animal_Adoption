```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
```

```python
df=pd.read_csv('../result.csv')
```

```python
df.head(10)
```

```python
lst_loc=[]
for i in df['found_location']:
    new=i.split(' in ')
    i=new[0]+','+new[1]
    lst_loc.append(i)
lst_loc
```

```python
df['found_location']=lst_loc
df.head()
```

```python
df_test=pd.DataFrame(data=df['found_location']).iloc[0:10000,:]
df_test
```

```python
address=df_test['found_location'].tolist()
address
```

```python
geolocator = Nominatim(user_agent="geoapiExercises")
```

```python
lst_new=[]
for i in address:
    location = geolocator.geocode(i)
    lst_new.append(location)
    print('Generating location: '+i)
    
print('-----------------------------------')
print("--------------Done-----------------")
lst_new
```

```python
len(lst_new)
```

```python
df_test=pd.DataFrame(data=df['found_location']).iloc[10000:20000,:]
df_test
```

```python
address=df_test['found_location'].tolist()
address
```

```python
for i in address:
    location = geolocator.geocode(i)
    lst_new.append(location)
    print('Generating location: '+i)
    
print('-----------------------------------')
print("--------------Done-----------------")
```

```python
len(lst_new)
```

```python
df_test=pd.DataFrame(data=df['found_location']).iloc[20000:,:]
address=df_test['found_location'].tolist()
address
```

```python
for i in address:
    location = geolocator.geocode(i)
    lst_new.append(location)
    print('Generating location: '+i)
    
print('-----------------------------------')
print("--------------Done-----------------")
```

```python
len(lst_new)
```

```python
count=0
for i in lst_new:
    if i==None:
        count+=1
count
```

```python
df['address']=lst_new
df.head()
```

```python
lat=[]
lon=[]
add_full=[]
zip_code=[]
for i in lst_new:
    if i==None:
        lat.append(None)
        lon.append(None)
        add_full.append(None)
        zip_code.append(None)
    else:
        lat.append(float(i.raw['lat']))
        lon.append(float(i.raw['lon']))
        add_full.append(i.raw['display_name'])
        zip_code.append(i.raw['display_name'].split(', ')[-2])
print(lat[:5],len(lat))
print(lon[:5],len(lon))
print(add_full[:5],len(add_full))
print(zip_code[:5],len(zip_code))
```

```python
df['lat']=lat
df['lon']=lon
df['address']=add_full
df['zip_code']=zip_code
df.head()
```

```python
new_zip=[]
for i in df['zip_code']:
    if len(str(i))==5 and i.isdigit() is True:
        i=i
    else:
        i=np.nan
    new_zip.append(i)
new_zip
```

```python
df['zip_code']=new_zip
```

```python
df.to_csv('../result_with_lat.csv', index=False)
```
