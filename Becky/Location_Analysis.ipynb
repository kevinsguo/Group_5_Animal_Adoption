```python
# import intake data
sql = '''
select * from animal_intake'''

intake_df = pd.read_sql(sql,con=connection)
intake_df.head()
```

```python
# import outcome data
sql = '''
select * from animal_outcome'''

outcome_df = pd.read_sql(sql,con=connection)
outcome_df.head()
```

```python
# merge data
df = intake_df.merge(outcome_df, on='animal_id', how='left')
df.head()
```

```python
found_location = df.found_location.value_counts()
```

```python
for i in found_location.index.tolist():
    print(i)
```

```python
df[["found_address", "found_city"]]= df["found_location"].str.split(" in", expand = True)
df.head()
```

```python
df[['found_city', 'found_state']] = df['found_city'].str.split('(', expand = True)
df.head()
```

```python
df.found_state = df.found_state.apply(lambda x: 'TX' if x == 'TX)' else None)
df.head()
```

```python
df.to_csv('results.csv', index=False, sep=',')
```
